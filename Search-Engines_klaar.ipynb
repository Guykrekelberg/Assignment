## Motivation.
We all use it on a daily basis: a search engine. We rely on it. Our guess is that you make your search queries with Google, just like a majority of the world population. Google’s global market share for desktop use has been rising from [68 until 78](https://www.netmarketshare.com/search-engine-market-share.aspx?qprid=5&qpcustomb=0 “Desktop Search Engine Market Share”) percent in 2016 alone, and in the mobile and tablet segment it has remained around a staggering [94](https://www.netmarketshare.com/search-engine-market-share.aspx?qprid=5&qpcustomb=1 “Mobile/Tablet Search Engine Market Share”) percent. These numbers tell us that Google has retrieved considerable market power and is close to monopolization. It is therefore not a surprise that European Commission has been investigating the tech giant. In a press release of July 2016 [two statements](http://europa.eu/rapid/press-release_IP-16-2532_en.htm) have come forward: 
> *“Google has abused its dominant position by systematically favouring its comparison shopping service in its search result pages”* and *“the company has abused its dominant position by artificially restricting the possibility of third party websites to display search advertisements from Google's competitors”*.

These results have been supported by [Patterson (2013)](http://ir.lawnet.fordham.edu/cgi/viewcontent.cgi?article=1514&context=faculty_scholarship) as well. Furthermore, [Pollock (2013)](https://mpra.ub.uni-muenchen.de/10369/2/MPRA_paper_10369.pdf “Is Google the next Microsoft?”) also shows that the welfare-optimizing search quality deviates from that of the profit-maximizing search quality. However, in light of these statements, is market dominance automatically a bad thing? It could it be a reflection of delivering higher search quality, in which case a monopoly is a reward to its honest efforts. Antitrust prosecutions would herewith disincentive superiority. But, as [Peter Norvig](https://www.theguardian.com/commentisfree/2015/apr/18/google-eu-monopoly-inquiry-too-late-to-stop), Google’s leading research guru, once put it: *”Google doesn’t have better algorithms – it just has more data”*. Being superior in search quality with increasing complexity algorithms is nowadays not the reason for market dominance. Therefore the statements by the EU of Google abusing its market power can hardly be defended. 

As you may have noticed, Google’s market power is under scrutiny and should be questioned. It is therefore of interest to see how it has developed until this point over time for it to make it understandable, but also to see how policies could influence market competition in benefit of consumers and total welfare. 

<a id='MethodSection'></a>

## Method.
In order to show that the development in the search engine market is worrying, we solve the [Argenton & Prüfer model](https://pure.uvt.nl/ws/files/1373523/search_engines.pdf "Our Inspiration") for each possible competition level inside the market. The three levels we consider are the competitive triopoly, the non-competitive triopoly and the duopoly. Market shares, profits, producer surplus, consumer surplus and total welfare are computed. 
Then we will show which policy needs to be implemented in order to make the search engine market healthy again, and compare this possible policy implementation to the current situation.
A reality check is done in order to prove how the market has been developing over time.
At last, a senstivity analysis will be executed to show the effect of an increase in advertising revenue on the market conditions.

<a id='QuickAnswer'></a>

## Quick Answer.
Welfare increases in $N_1$ (- *over time*) in the three modelled structures Duopoly, Triopoly and Competitive Triopoly. However, in the Duopoly and Triopoly the welfare growth slows down and in the Competitive Triopoly it monotonically increases. In the latter case, which is the policy recommendation, the increased competition also delivers higher search quality, rate of innovation and consumer surplus (only producer surplus stagnates). 

<a id='MainAssumptions'></a>

## Main Assumptions. 
1. The advertising revenue of a search engine is proportional to its market share, which is broadly in line with industry’s figures. This is also supported with advertisers’ [reasoning]( http://www.investopedia.com/articles/investing/060315/why-facebook-biggest-threat-googles-adrevenue-throne.asp "Investopedia - Facebook threats Google in Ad Revenue") where the amount of visitors increases its demand. 
2. Consumers can only rationally expect search engine quality to be different on average, for they do not know which engine delivers the best quality for each query. Therefore the firm's market share is monotonically increasing in its perceived average search quality. 
3. There is a unit mass of consumers, each of which has demand for one query, and the cost for a consumers to use a search engine is zero.
4. Moreover, the quality is high enough such that everyone will use a search engine, meaning that the market is fully covered.

### Introducing Model Characteristics

$x_i$ = the simultaneous search quality of firm $i$

$N_i$ = the installed base of firm $i$, in other words it is the amound of previous search queries run

$F$ = the fixed cost for operations

$p$ = the exogenously given advertising revenue associated to one consumer

$D_i = \frac{x_i}{\sum_{i=1}^{n} x_i}$ = the market share of firm $i$

$C(x_i) = \frac{x_i}{N_i}$ = the costs of production of quality $x_i$

The maximization problem each firm will solve is the following:

$\underset{x_i}{\text{max}} \hspace{10pt} \Pi_i = \frac{x_i}{\sum_{i=1}^{n} x_i}p - \frac{x_i}{N_i} - F$

<a id='Triopoly'></a>
## Triopoly

Since the search engine market is a triopoly at first, where the model is a static game with complete information, the (Nash) Equilibrium values are obtained by solving the set of reaction functions $x_i(p, N, \sum{x_j})$. Where the equilibrium values $x_i^*$ are as follows:

$
\begin{align}
x_1 & = \sqrt{pN_1(x_2 + x_3)} - x_2 - x_3 
&& x_2 = \sqrt{pN_2(x_1 + x_3)} - x_1 - x_3 
&&& x_3 = \sqrt{p(x_1 + x_2)} - x_1 - x_2 \\
x_1^* & = \sqrt{\frac{p^2N^2_1N^2_2}{(N_2+N_1(1+N_2))^2}} - \frac{4pN_1N^2_2}{(N_2+N_1(1+N_2))^2}
&& x_2^* = \frac{2pN_1N_2(N_1(N_2-1)+N_2)}{(N_2+N_1(1+N_2))^2}
&&& x_3^* = \frac{2pN_1N_2(-N_1(N_2-1)+N_2)}{(N_2+N_1(1+N_2))^2}\\
\end{align}
$

The python code to show the development of $x_i^*$ over $N_1$ is plotted below, where $N_2$ is taken as a constant *($N_3$ is irrelevant for the equilibrium)*, $p$ is set to 1 and fixed costs are considered nil:

```python
def investment_trio_1 (N_1, N_2, p):
    return (((p**2)*(N_1)**2*(N_2)**2)/(N_2+N_1*(1+N_2))**2)**(1/2)- (4*p*N_1*N_2**2)/((N_2+N_1*(1+N_2))**2)
def investment_trio_2 (N_1, N_2, p):
    return 2*p*N_1*N_2*(N_1*(N_2-1)+N_2)/((N_2+N_1*(1+N_2))**2)
def investment_trio_3 (N_1, N_2, p):
    return 2*p*N_1*N_2*(-N_1*(N_2-1)+N_2)/((N_2+N_1*(1+N_2))**2)

N_2 = 1.2 ## just an example (same as in paper)
p = 1.0
F=0

range_N_1 = arange(1.20,10.1,0.1)

range_investment_trio_1= [investment_trio_1 (N_1, N_2, p) for N_1 in range_N_1]
range_investment_trio_2= [investment_trio_2 (N_1, N_2, p) for N_1 in range_N_1]
range_investment_trio_3= [investment_trio_3 (N_1, N_2, p) for N_1 in range_N_1]


plt.clf()
plt.plot(range_N_1,range_investment_trio_1,'-', color = 'r', linewidth = 2)
plt.plot(range_N_1,range_investment_trio_2,'-', color = 'b', linewidth = 2)
plt.plot(range_N_1,range_investment_trio_3,'-', color = 'g', linewidth = 2)
plt.ylim (0.0, 1.0)
plt.xlim (1.2, 10.0)
plt.title("Equilibrium Search Quality Triopoly",fontsize = 15)
plt.xlabel("$N_1$",fontsize = 15)
plt.ylabel("$x_i$",fontsize = 15,rotation = 90)
plt.annotate('$Firm 1$', xy=(2.5,investment_trio_1(2.5,N_2,p)), xycoords='data', 
              xytext=(30, 50), textcoords='offset points', size = 20,
              arrowprops=dict(arrowstyle="simple", facecolor='r', linewidth = 2,
                              connectionstyle="arc3,rad=.2"),
              )
plt.annotate('$Firm 2$', xy=(2.7,0.24),  xycoords='data', 
              xytext=(30, 50), textcoords='offset points', size = 20,
              arrowprops=dict(arrowstyle="simple", facecolor='b', linewidth = 2,
                              connectionstyle="arc3,rad=.2"),
              )
plt.annotate('$Firm 3$', xy=(4.3,0.05),  xycoords='data', 
              xytext=(30, 50), textcoords='offset points', size = 20,
              arrowprops=dict(arrowstyle="simple", facecolor='g', linewidth = 2,
                              connectionstyle="arc3,rad=.2"),
              )

```

As we can see, a clear divergence happens as the installed base of firm 1 $N_1$ increases. The search quality $x_1$ increases due to more query logs in their search engine and the competitors' quality suffer from their loss in search queries run on their engine. So, that $x_2$ and $x_3$ both decrease, but since firm 3 started with the lowest quality they are hit the most and will leave the market when $N_1$ gets big enough and their own quality diminishes. 

The market shares $D_i^* = \frac{x_i^*}{\sum_{i=1}^{n} x_i^*}$ develop the same way as $x_i^*$, since recall that the market share is monotonically increasing in its perceived average search quality. The Nash equilibrium market shares are:

$
\begin{align}
& D_1^* = 1 - \frac{2N_2}{N_2 + N_1(1+N_2)}
&& D_2^* = 1 - \frac{2N_1}{N_2 + N_1(1+N_2)}
&&& D_3^* = \frac{N_1(1-N_2)+N_2}{N_2+N_1(1+N_2)}
\end{align}
$

In Python code, this translates to the following, where we use the same trick but different functions:

```python
def marketshare_1(N_1,N_2):
    return (1-(2*N_2)/(N_2+N_1*(1+N_2)))
def marketshare_2(N_1,N_2):
    return (1-(2*N_1)/(N_2+N_1*(1+N_2)))
def marketshare_3(N_1,N_2):
    return (N_1*(1-N_2)+N_2)/(N_2+N_1*(1+N_2))


range_marketshare_1= [marketshare_1 (N_1,N_2) for N_1 in range_N_1]
range_marketshare_2= [marketshare_2 (N_1,N_2) for N_1 in range_N_1]
range_marketshare_3= [marketshare_3 (N_1,N_2) for N_1 in range_N_1]

plt.clf()
plt.plot(range_N_1,range_marketshare_1,'-', color = 'r', linewidth = 2)
plt.plot(range_N_1,range_marketshare_2,'-', color = 'b', linewidth = 2)
plt.plot(range_N_1,range_marketshare_3,'-', color = 'g', linewidth = 2)
plt.ylim (0.0, 1.0)
plt.xlim (1.2, 10.0)
plt.title("Equilibrium Market Shares Triopoly",fontsize = 15)
plt.xlabel("$N_1$",fontsize = 15)
plt.ylabel("$D_i$",fontsize = 15,rotation = 90)
plt.annotate('$Firm 1$', xy=(2.8,0.7), xycoords='data', 
              xytext=(30, 50), textcoords='offset points', size = 20,
              arrowprops=dict(arrowstyle="simple", facecolor='r', linewidth = 2,
                              connectionstyle="arc3,rad=.2"),
              )
plt.annotate('$Firm 2$', xy=(2.9,0.24), xycoords='data', 
              xytext=(30, 50), textcoords='offset points', size = 20,
              arrowprops=dict(arrowstyle="simple", facecolor='b', linewidth = 2,
                              connectionstyle="arc3,rad=.2"),
              )
plt.annotate('$Firm 3$', xy=(5,0.02), xycoords='data', 
              xytext=(30, 50), textcoords='offset points', size = 20,
              arrowprops=dict(arrowstyle="simple", facecolor='g', linewidth = 2,
                              connectionstyle="arc3,rad=.2"),
              )

```

Also in the market shares values we notice firm 3 leaving the market, due to their continuous decrease of search quality due to the loss of their installed base. Firm 2 still survives, but as we will see in the duopoly case, the situation and optimization problems are altered due to the different market structure. Furthermore, the dominance of firm 1 becomes more and more apparent. 

Now that we have determined $D_i^*$ and $x_i*$, and for a given $N_1$, $N_2$, $F$, and $p$, we can determine and plot the equilibrium profits as given at the <a href='#ModelCharacteristics'>Model Characteristics</a>. We get:

$
\begin{align}
& \Pi_1^* = \frac{(N_2-N_1(1+N_2))^2}{(N_2+N_1(1+N_2))^2}p-F
&& \Pi_2^* = \frac{(N_2+N_1(N_2-1))^2}{(N_2+N_1(1+N_2))^2}p-F
&&& \Pi_3^* = \frac{(N_2+N_1(1-N_2))^2}{(N_2+N_1(1+N_2))^2}p-F
\end{align}
$

In Python code the plot of equilibrium profits over $N_1$ is obtained as follows:

```python
def profit_1(N_1,N_2,p,F):
    return ((N_2-N_1*(1+N_2))**2)/(N_2+N_1*(1+N_2))**2*p-F
def profit_2(N_1,N_2,p,F):
    return ((N_2+N_1*(N_2-1))**2)/(N_2+N_1*(1+N_2))**2*p-F
def profit_3(N_1,N_2,p,F):
    return ((N_2+N_1*(1-N_2))**2)/(N_2+N_1*(1+N_2))**2*p-F


range_profit_1= [profit_1(N_1,N_2,p,F) for N_1 in range_N_1]
range_profit_2= [profit_2(N_1,N_2,p,F) for N_1 in range_N_1]
range_profit_3= [profit_3(N_1,N_2,p,F) for N_1 in range_N_1]

plt.clf()
plt.plot(range_N_1,range_profit_1,'-', color = 'r', linewidth = 2)
plt.plot(range_N_1,range_profit_2,'-', color = 'b', linewidth = 2)
plt.plot(range_N_1,range_profit_3,'-', color = 'g', linewidth = 2)
plt.ylim (0.00000001, 1.0)
plt.xlim (1.2, 10.0)
plt.title("Equilibrium Profits Triopoly",fontsize = 15)
plt.xlabel("$N_1$",fontsize = 15)
plt.ylabel("$\pi_i$",fontsize = 15,rotation = 90)
plt.annotate('$Firm 1$', xy=(2.9,0.5), xycoords='data', 
              xytext=(30, 50), textcoords='offset points', size = 20,
              arrowprops=dict(arrowstyle="simple", facecolor='r', linewidth = 2,
                              connectionstyle="arc3,rad=.2"),
              )
plt.annotate('$Firm 2$', xy=(6,0.05), xycoords='data', 
              xytext=(30, 50), textcoords='offset points', size = 20,
              arrowprops=dict(arrowstyle="simple", facecolor='b', linewidth = 2,
                              connectionstyle="arc3,rad=.2"),
              )
plt.annotate('$Firm3$', xy=(2,0.03), xycoords='data', 
              xytext=(30, 50), textcoords='offset points', size = 20,
              arrowprops=dict(arrowstyle="simple", facecolor='g', linewidth = 2,
                              connectionstyle="arc3,rad=.2"),
              )
```

Just like in the previous two graphs, we notice the same divergence, but then in profits. Firm 1 clearly benefits from their market share increase and firm 2 and 3 suffer. Yet again, firm 2 manages to keep making a small amount of profits. 

However, to see whether the modelled development in a triopoly case benefits society, besides only firms' profits, we will have a look at welfare and distinguish between producer and consumer surplus, where welfare = producer surplus + consumer surplus:

$ Producer \, Surplus = \Pi_1 + \Pi_2 + \Pi_3 = \dfrac{(3N_2^2 - 2N_1N_2(1+N_2)+N_1^2(3+N_2(3N_2-2)))}{N_2+N_1(1+N_2))^2}p-3F $

$ Consumer \, Surplus = D_1x_1^*+D_2x_2^*+D_3x_3^* = \dfrac{2N_1N_2(3N_2^2-2N_1N_2(1+N_2)+N_1^2(3+N_2(3N_2-2))}{N_2+N_1(1+N_2))^3}p $

$ Welfare = PS + CS = \dfrac{(N_2+N_1(3N_2+1))(3N_2^2-2N_1N_2(1+N_2)+N_1^2(3+N_2(3N_2-2))}{N_2+N_1(1+N_2))^3}p-3F $

Solving for these equations over N translates into the following code and graph:

```python
##Welfare

def producersurplus_triopoly(N_1,N_2,p,F):
    return ((3*N_2-2*N_1*N_2*(1+N_2)+N_1**2*(3+N_2*(3*N_2-2)))*p/(N_2+N_1*(1+N_2))**2-3*F)

def consumersurplus_triopoly(N_1,N_2,p):
    return (2*N_1*N_2*(3*N_2**2-2*N_1*N_2*(1+N_2)+N_1**2*(3+N_2*(3*N_2-2))))*p/(N_2+N_1*(1+N_2))**3

def welfare_triopoly(N_1,N_2,p,F):
    return (((N_2+N_1*(3*N_2+1))*(3*(N_2**2)-2*N_1*N_2*(1+N_2)+(N_1**2)*(3+N_2*(3*N_2-2)))*p)/(N_2+N_1*(1+N_2))**3)-3*F

range_PS= [producersurplus_triopoly(N_1,N_2,p,F) for N_1 in range_N_1]
range_CS= [consumersurplus_triopoly(N_1,N_2,p) for N_1 in range_N_1]
range_Welfare= [welfare_triopoly(N_1,N_2,p,F) for N_1 in range_N_1]


plt.clf()
plt.plot(range_N_1,range_PS,'-', color = 'r', linewidth = 2)
plt.plot(range_N_1,range_CS,'-', color = 'b', linewidth = 2)
plt.plot(range_N_1,range_Welfare,'-', color = 'g', linewidth = 2)
plt.ylim (0.00000001, 1.6)
plt.xlim (1.2, 10.0)
plt.title("Welfare Triopoly",fontsize = 15)
plt.xlabel("$N_1$",fontsize = 15)
plt.ylabel("$welfare$",fontsize = 15,rotation = 90)
plt.annotate('$Welfare$', xy=(3.2,1.1), xycoords='data', 
              xytext=(30, 50), textcoords='offset points', size = 20,
              arrowprops=dict(arrowstyle="simple", facecolor='g', linewidth = 2,
                              connectionstyle="arc3,rad=.2"),
              )
plt.annotate('$CS$', xy=(8,0.86), xycoords='data', 
              xytext=(30, 50), textcoords='offset points', size = 20,
              arrowprops=dict(arrowstyle="simple", facecolor='b', linewidth = 2,
                              connectionstyle="arc3,rad=.2"),
              )
plt.annotate('$PS$', xy=(2.9,0.53), xycoords='data', 
              xytext=(30, 50), textcoords='offset points', size = 20,
              arrowprops=dict(arrowstyle="simple", facecolor='r', linewidth = 2,
                              connectionstyle="arc3,rad=.2"),
              )
```

Fortunately, besides increasing profits, the consumer surplus isn't harmed and increases with $N_1$. The welfare is increasing as the market structure develops and shows that society will benefit in the triopoly cases with the current market architecture. What is interesting to see is that consumer surplus overtakes producer surplus at around $N_1 = 6$, which you wouldn't expect since Firm 1 gains market share and market power and should be able to accrue more of the benefits to themselves. An explanation is that the loss of profits from the other firms is more than what firm 1 gains. However the story isn't finished, since a triopoly is not sustainable and we move on to the duopoly case as shown in the graph portraying the market shares.

# __Duopoly__
Over time in the triopoly case, a firm will not be able to compete anymore and make profits. In that instance he will leave the market and a duopoly scenario exists. In this part we will graphically show what happens in the duopoly case. The model assumptions, characteristics and profit maximization problem remain untouched, however there will only be two reaction function to solve for the Nash Equilibrium $x_i^*$ and $D_I^*$:

$
\begin{align}
& x_1^* = \frac{N_1^2}{(1+N_1)^2} && x_2^* = \frac{N_1}{(1+N_1)^2} \\
& D_1^* = \frac{N_1}{1+N_1} && D_2^* = \frac{1}{1+N_1}
\end{align}
$

This is codified in the following manner:

```python
## duopoly
def investment_reaction_function (N_1):
    return (N_1**2)/(1+N_1)**2
def investment_reaction_2(N_1):
    return N_1/(1+N_1)**2

range_N_1 = arange(1.20,10.1,0.1)
range_investment_reaction_function_1= [investment_reaction_function(N_1) for N_1 in range_N_1]
range_investment_reaction_2= [investment_reaction_2(N_1) for N_1 in range_N_1]

plt.clf()
plt.plot(range_N_1,range_investment_reaction_2,'-', color = 'b', linewidth = 2)
plt.plot(range_N_1,range_investment_reaction_function_1,'-', color = 'r', linewidth = 2)
plt.title("Equilibrium Search Quality Levels Duopoly",fontsize = 15)
plt.xlim (1.2, 10.0)
plt.xlabel("$N_1$",fontsize = 15)
plt.ylabel("$x_i$",fontsize = 15,rotation = 90)
plt.annotate('$Firm 1$', xy=(2.5,0.53), xycoords='data', 
              xytext=(30, 50), textcoords='offset points', size = 20,
              arrowprops=dict(arrowstyle="simple", facecolor='r', linewidth = 2,
                              connectionstyle="arc3,rad=.2"),
              )
plt.annotate('$Firm 2$', xy=(5,0.14), xycoords='data', 
              xytext=(30, 50), textcoords='offset points', size = 20,
              arrowprops=dict(arrowstyle="simple", facecolor='b', linewidth = 2,
                              connectionstyle="arc3,rad=.2"),
              )
```

The development of the search qualities, again over $N_1$, shows a somewhat same path as in the triopoly case. The one firm that leads in the beginning sees his quality increase at the cost of the other firm(s). However, in duopoly, the search quality does not reach zero as soon as in the triopoly case. 

```python
def marketshare_duo_1(N_1):
    return (N_1)/(1+N_1)
def marketshare_duo_2(N_1):
    return (1/(N_1+1))
            

range_marketshare_duo_1= [marketshare_duo_1(N_1) for N_1 in range_N_1]
range_marketshare_duo_2= [marketshare_duo_2(N_1) for N_1 in range_N_1]


plt.clf()
plt.plot(range_N_1,range_marketshare_duo_1,'-', color = 'r', linewidth = 2)
plt.plot(range_N_1,range_marketshare_duo_2,'-', color = 'b', linewidth = 2)
plt.ylim (0.0, 1.0)
plt.xlim (1.2, 10.0)
plt.title("Equilibrium Market Shares Duopoly",fontsize = 15)
plt.xlabel("$N_1$",fontsize = 15)
plt.ylabel("$D_i$",fontsize = 15,rotation = 90)
plt.annotate('$Firm 1$', xy=(2,0.67), xycoords='data', 
              xytext=(30, 50), textcoords='offset points', size = 20,
              arrowprops=dict(arrowstyle="simple", facecolor='r', linewidth = 2,
                              connectionstyle="arc3,rad=.2"),
              )
plt.annotate('$Firm 2$', xy=(5.9,0.15), xycoords='data', 
              xytext=(30, 50), textcoords='offset points', size = 20,
              arrowprops=dict(arrowstyle="simple", facecolor='b', linewidth = 2,
                              connectionstyle="arc3,rad=.2"),
              )
```

Like the development of the search qualities, the market share paths diverge in benefit of the leading firm 1 and at the cost of firm 2. The market share of firm 2 is converging to zero, but is not noticable in the graph. This shows that it will take more time for the competitor of the leading firm to leave the market in the duopoly case. 

Furthermore, in the same way as in the Triopoly Case, we can obtain the equilibrium profits given $x_i^*$ and $D_i^*$ with substitution:
$ \Pi_1 = \frac{N_1^2}{(1+N_1)^2}-F ,\hspace{40pt} \Pi_2 = \frac{1}{(1+N_1)^2}-F $

The graph is produced by the following:

```python
def profit_duo_1(N_1,F):
    return ((N_1)**2)/(1+N_1)**2-F
def profit_duo_2(N_1,F):
    return 1/(1+N_1)**2-F

range_N_1 = arange(1.20,10.1,0.1)

range_profit_duo_1= [profit_duo_1(N_1,F) for N_1 in range_N_1]
range_profit_duo_2= [profit_duo_2(N_1,F) for N_1 in range_N_1]

plt.clf()
plt.plot(range_N_1,range_profit_duo_1,'-', color = 'r', linewidth = 2)
plt.plot(range_N_1,range_profit_duo_2,'-', color = 'b', linewidth = 2)
plt.ylim (0.00000001, 1.0)
plt.xlim (1.2, 10.0)
plt.title("Equilibrium Profits Duopoly",fontsize = 15)
plt.xlabel("$N_1$",fontsize = 15)
plt.ylabel("$\pi_i$",fontsize = 15,rotation = 90)
plt.annotate('$Firm 1$', xy=(2.9,0.57), xycoords='data', 
              xytext=(30, 50), textcoords='offset points', size = 20,
              arrowprops=dict(arrowstyle="simple", facecolor='r', linewidth = 2,
                              connectionstyle="arc3,rad=.2"),
              )
plt.annotate('$Firm 2$', xy=(4.5,0.05), xycoords='data', 
              xytext=(30, 50), textcoords='offset points', size = 20,
              arrowprops=dict(arrowstyle="simple", facecolor='b', linewidth = 2,
                              connectionstyle="arc3,rad=.2"),
              )
```

The profits in the equilibrium reveal the same situation as before and as expected from the market shares and search quality. Firm 1 gains from their dominant and leading position. 

If we then have a look at the total welfare, producer surplus and consumer surplus, we can see whether a duopoly case is beneficial over time for the whole society. The equations that hold are:

$ Producer \, Surplus = \dfrac{N_1^2+1}{(1+N_1)^2}-2F $

$ Consumer \, Surplus = \dfrac{N_1+N_1^3}{(1+N_1)^3} $

$ Total \, Welfare = \dfrac{1+2N_1+N_1^2+2N_1^3}{(1+N_1)^3}-2F $

Computed in Python by:

```python
##Welfare

def producersurplus_duopoly(N_1,F):
    return (N_1**2+1)/(1+N_1)**2-2*F

def consumersurplus_duopoly(N_1):
    return (N_1+N_1**3)/(1+N_1)**3

def welfare_duopoly(N_1, F):
    return (1+2*N_1+N_1**2+2*N_1**3)/(1+N_1)**3-2*F

range_PS_duo= [producersurplus_duopoly(N_1,F) for N_1 in range_N_1]
range_CS_duo= [consumersurplus_duopoly(N_1) for N_1 in range_N_1]
range_Welfare_duo= [welfare_duopoly(N_1,F) for N_1 in range_N_1]


plt.clf()
plt.plot(range_N_1,range_PS_duo,'-', color = 'r', linewidth = 2)
plt.plot(range_N_1,range_CS_duo,'-', color = 'b', linewidth = 2)
plt.plot(range_N_1,range_Welfare_duo,'-', color = 'g', linewidth = 2)
plt.ylim (0.00000001, 1.6)
plt.xlim (1.2, 10.0)
plt.title("Welfare Duopoly",fontsize = 15)
plt.xlabel("$N_1$",fontsize = 15)
plt.ylabel("$welfare$",fontsize = 15,rotation = 90)
plt.annotate('$Welfare$', xy=(2.9,1.085), xycoords='data', 
              xytext=(30, 50), textcoords='offset points', size = 20,
              arrowprops=dict(arrowstyle="simple", facecolor='g', linewidth = 2,
                              connectionstyle="arc3,rad=.2"),
              )
plt.annotate('$PS$', xy=(4.5,0.7), xycoords='data', 
              xytext=(30, 50), textcoords='offset points', size = 20,
              arrowprops=dict(arrowstyle="simple", facecolor='r', linewidth = 2,
                              connectionstyle="arc3,rad=.2"),
              )
plt.annotate('$CS$', xy=(2.7,0.44), xycoords='data', 
              xytext=(30, 50), textcoords='offset points', size = 20,
              arrowprops=dict(arrowstyle="simple", facecolor='b', linewidth = 2,
                              connectionstyle="arc3,rad=.2"),
              )
```

# Policy Recommendation
In contrast to the previous cases, in which $N_i \neq N_j$, the policy which is recommended is that the querie logs amongst all search engines are shared such that $N_1 = N_j$. That is, the installed bases are equal and we have $N = N_1 + N_2 + N_3$. This situation is dubbed a Competitive Triopoly, since the tipping of the market is not due to the differences in installed bases anymore. The market has become competitive in $x_i$ only. 

The Nash Equilibrium values for the quality level is: $x_1^* = x_2^* = x_3^* = \tfrac{2}{9}Np = x^{competitive}$. Where this results in equal market share of one-third for each. The equilibrium profits, consumer surplus, producer surplus and welfare are given as:

$\Pi_1 = \Pi_2 = \Pi_3 = \tfrac{p}{9}-F $

$Producer \, Surplus = \tfrac{p}{3} - 3F $

$Consumer \, Surplus = \tfrac{2}{9}Np $

$and \, Welfare = \tfrac{p}{9}(3+2N)-3F $

The graphs of Quality Levels/Equilibrium Investments and that of Welfare have been computed and shown below:

```python
def investment_triopoly(N_1,p):
    return (2.0/9*N_1*p)


range_investment_triopoly= [investment_triopoly(N_1,p) for N_1 in range_N_1]

plt.clf()
plt.plot(range_N_1,range_investment_triopoly,'-', color = 'r', linewidth = 2)
plt.ylim (0.0, 2.0)
plt.xlim (1.2, 10.0)
plt.title("Equilibrium  Search Quality Levels Comp. Triopoly",fontsize = 15)
plt.xlabel("$N_1$",fontsize = 15)
plt.ylabel("$x_i$",fontsize = 15,rotation = 90)
plt.annotate('$Firm 1&2&3$', xy=(4.7,0.7), xycoords='data', 
              xytext=(30, 50), textcoords='offset points', size = 20,
              arrowprops=dict(arrowstyle="simple", facecolor='r', linewidth = 2,
                              connectionstyle="arc3,rad=-1"),
              )
```

As we see, and can deduce from $x_i = \frac{2}{9}Np$, the simultaneous search quality increases with the querie logs available. However, since the querie logs are shared amongst all firms in this new market structure, the $N_1 = N_j$ does not indirectly refer to the factor time per sé. In previous cases this could be linked, but a changing market share doesn't affect the installed database anymore and recall that the market shares are even fixed to one-third. 

```python
def producersurplus_comp_triopoly(p,F):
    return p*(1.0/3)- 3*F

def consumersurplus_comp_triopoly(N_1,p):
    return (2.0/9)*N_1*p

def welfare_comp_triopoly(N_1,p,F):
    return p*(1.0/9)*(3+2*N_1)-3*F


range_PS_comp_trio= [producersurplus_comp_triopoly(p,F) for N_1 in range_N_1]
range_CS_comp_trio= [consumersurplus_comp_triopoly(N_1,p) for N_1 in range_N_1]
range_Welfare_comp_trio= [welfare_comp_triopoly(N_1,p,F) for N_1 in range_N_1]


plt.clf()
plt.plot(range_N_1,range_PS_comp_trio,'-', color = 'r', linewidth = 2)
plt.plot(range_N_1,range_CS_comp_trio,'-', color = 'b', linewidth = 2)
plt.plot(range_N_1,range_Welfare_comp_trio,'-', color = 'g', linewidth = 2)
plt.ylim (0.00000001, 1.6)
plt.xlim (1.2, 10.0)
plt.title("Welfare Comp. Triopoly",fontsize = 15)
plt.xlabel("$N_1$",fontsize = 15)
plt.ylabel("$welfare$",fontsize = 15,rotation = 90)
plt.annotate('$PS$', xy=(5.9,0.34), xycoords='data', 
              xytext=(30, 50), textcoords='offset points', size = 20,
              arrowprops=dict(arrowstyle="simple", facecolor='r', linewidth = 2,
                              connectionstyle="arc3,rad=.2"),
              )
plt.annotate('$CS$', xy=(4.7,0.8), xycoords='data', 
              xytext=(30, 50), textcoords='offset points', size = 20,
              arrowprops=dict(arrowstyle="simple", facecolor='b', linewidth = 2,
                              connectionstyle="arc3,rad=-0.8"),
              )
plt.annotate('$Welfare$', xy=(1.5,1), xycoords='data', 
              xytext=(30, 50), textcoords='offset points', size = 20,
              arrowprops=dict(arrowstyle="simple", facecolor='g', linewidth = 2,
                              connectionstyle="arc3,rad=.7"),
              )
```

The Competitive Triopoly in CS, PS and Welfare terms seem to show a different story than in the Triopoly and Duopoly case. Producer surplus is unaffected due to the increased competition for any $N_1 = N_j$ and consumers will benefit from a bigger dataset of querie logs due to better search qualities. Welfare is thus mainly determined by the effect of $N$ on consumer surplus.
